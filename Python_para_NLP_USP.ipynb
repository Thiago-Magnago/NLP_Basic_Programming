1.Dado o arquivo qbdata.txt, retorne o rating de cada QB na forma “nome do QB teve valor XX.X’ e escreva em um arquivo novo.

```python
#Vamos criar um arquivo de entrada (qbdata) e de saída
infile = open('/content/drive/MyDrive/Python_para_NLP_USP/qbdata.txt', 'r')
outfile = open('/content/drive/MyDrive/Python_para_NLP_USP/result_qbdata.txt', 'w')
```

```python
conteudo = infile.read()
print(conteudo)
```

```python
#Quebrando as linhas em elementos de uma lista, a partir da quebra de linha '\n'.
lista = (conteudo.split('\n'))
lista
```

```python
#Quebrando os elementos na vírgula, de modo a obter listas onde se possa captar o nome
lista_de_listas = [item.split(',') for item in lista]
lista_de_listas
```

```python
#Criando uma lista apenas com os nomes
nomes_qb = [item[0] for item in lista_de_listas]
nomes_qb
```

```python
#Criando uma lista tokenizada, apenas com os elementos complementares (exceto nome)...
#...de modo a permitir a captura do último elemento da lista.
complemento_qb = [item[1].split() for item in lista_de_listas]
complemento_qb
```

```python
#Criando uma lista apenas com os elementos de interesse (últimos da lista anterior)
ultima_qb = [item[-1] for item in complemento_qb]
ultima_qb
```

```python
#Laço para juntar nome e nota, com o texto pedido pela questão.
texto_pronto=[]
for nome, nota in zip(nomes_qb, ultima_qb):
  texto = nome + ' teve nota ' + nota
  texto_pronto.append(texto)
print(texto_pronto)
```

```python
#Convertendo a lista de strings para uma string, separando-os por quebra de linha '\n'.
texto_final = "\n".join(texto_pronto)
print(texto_final)
```

```python
#Escrever o texto final no arquivo criado (representado na variável 'outfile') e fechá-lo
outfile.write(texto_final)
outfile.close()
```

Tokenizar sem pontuação: trazer palavras, números e underscore (_).

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+') #criando meu tokenizador. O RegexpTokenizer recebe uma Regex que dirá o padrão de tokens que procuro.
#Esta Regex retornará: todas as palavras, números e underscore do texto (_), ignorando todas as pontuações.
#Dica para teste da Regex: pythex.org
tokens = tokenizer.tokenize(texto) #chamamos o objeto 'tokenizer', que tem o método 'tokenize'.
print(tokens)
```

Podemos melhorar o contador de frequencia normalizando todas as palavras para MINÚSCULAS.

Existem 2 formas: Uma mais braçal e outra mais avançada (list comprehension)

```python
#FORMA 1 (BRAÇAL)
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+') #Quero todos os tokens que contenham esses caracteres: A a Z minúsculo ou maiúsculo.
tokens = tokenizer.tokenize(corpus) #chamamos o objeto 'tokenizer', que tem o método 'tokenize'.

#Transformando os tokens em minúsculo numa nova lista
nova_lista = []
for token in tokens:
  nova_lista.append(token.lower())

#Contando a frequencia da nova lista
frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common()
```

Remoção de Stopwords - LOOP FOR

```python
from nltk.tokenize import RegexpTokenizer
stopwords = nltk.corpus.stopwords.words('portuguese')

#Tokenização
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+') #Quero todos os tokens que contenham esses caracteres: A a Z minúsculo ou maiúsculo.
tokens = tokenizer.tokenize(corpus) #chamamos o objeto 'tokenizer', que tem o método 'tokenize'.

#Transformando os tokens em MINÚSCULO numa nova lista, desde que o token não seja STOPWORD.
nova_lista = []

for token in tokens:
  if token.lower() not in stopwords:
    nova_lista.append(token.lower())

#Contando a frequencia da nova lista
frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common()
```

Remoção de Stopwords - LIST COMPREHENSION

```python
from nltk.tokenize import RegexpTokenizer
stopwords = nltk.corpus.stopwords.words('portuguese')

#Tokenização
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+') #Quero todos os tokens que contenham esses caracteres: A a Z minúsculo ou maiúsculo.
tokens = tokenizer.tokenize(corpus) #chamamos o objeto 'tokenizer', que tem o método 'tokenize'.

#Transformando os tokens em MINÚSCULO numa nova lista, desde que o token não seja STOPWORD.
'''nova_lista = []

for token in tokens:
  if token.lower() not in stopwords:
    nova_lista.append(token.lower())'''

nova_lista = [token.lower() for token in tokens if token.lower() not in stopwords]

#Contando a frequencia da nova lista
frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common()
```

QUAL A UTILIDADE DOS BIGRAMS, TRIGRAMS E NGRAMS?

Reconhecimento de entidades nomeadas. Nos bigrams, existem algumas palavras nas quais ambas começam com letra maiúscula. E como esse texto é bem 'comportado', podemos usar essa ideia. Podemos fazer algo do tipo: Todas as palavras que começarem com letras maiúsculas consideraremos como entidades nomeadas.

Ex. de entidade no bigram: ('Eli', 'Manning');

Ex. de entidade no trigram: ('Super', 'Bowl', 'XLII');

Reconhecimento de entidades nomeadas

```python
from nltk import bigrams
from nltk import trigrams

bigramas = list(bigrams(tokens))
trigramas = list(trigrams(tokens))
                        
#Iremos reconhecer entidades com a premissa: Ambas as letras iniciais maiúsculas
#Vamos iterar sobre as listas de tuplas, e verificarei elemento por elemento..
#...se a letra inicial é maiúscula, pelo método 'isupper()'.
#OBS.: Veremos que no SPACY esse processo é mais simples.

entidades_bigramas = []
for bigrama in bigramas:
  first_letter = bigrama[0][0]
  second_letter = bigrama[1][0]
  if first_letter.isupper() and second_letter.isupper():
    entidades_bigramas.append(bigrama)
  
entidades_trigramas = []
for trigrama in trigramas:
  first_letter = trigrama[0][0]
  second_letter = trigrama[1][0]
  third_letter = trigrama[2][0]
  if first_letter.isupper() and second_letter.isupper() and third_letter.isupper():
    entidades_trigramas.append(trigrama)
```

NLTK STEMMER e LEMMATIZER

Técnicas que propõem a desconsiderar a flexão das palavras, reduzindo as flexões a uma palavra só.

STEMMING - consiste em reduzir a palavras ao seu radical. (É uma técnica bem radical!)

- amig: amigo, amiga, amigão

- gat: gato, gata, gatos

- prop: propõem, propuseram, propondo


LEMATIZAÇÃO - consiste em reduzir a palavra à sua forma canônica, levando em conta sua classe gramatical.

- propor: propõem, propuseram, propondo

- estudar: estudando, estudioso, estudei


NLTK tem vários algoritmos para STEMMING:

- RSLP - Removedor de Sufixos da Lingua Portuguesa

- Porter

- ISRI

- Lancaster

- Snowball

Cada um deles é um algoritmo, treinado de uma forma específica. Cabe ao analista experimentar, testar e verificar o que se encaixa melhor para a tarefa em questão.

```python
#RSLP - Removedor de Sufixos da Lingua Portuguesa
#Mais detalhes sobre RSLP: http://www.inf.ufrgs.br/~viviane/rslp/index.htm
import nltk
stemmer = nltk.RSLPStemmer()
print(stemmer.stem('amigão'))
print(stemmer.stem('amigo'))
print(stemmer.stem('propuseram'))
print(stemmer.stem('propõem'))
print(stemmer.stem('propondo'))
print(stemmer.stem('estacionamento'))
print(stemmer.stem('estacionar'))
```

NLTK - ETIQUETADORES

O NLTK possui 2 corpus que servem como base para o etiquetador em **português**:

- Floresta (9.266 sentenças etiquetadas);

- Mac Morpho (51.397 sentenças etiquetadas).

Para o **inglês**, já existe um etiquetador padrão treinado:

- nltk.pos_tag()

Os etiquetadores passam primeiro por uma fase de treinamento com as sentenças presentes.

Como resultado, os etiquetadores retornam uma tupla ('palavra', 'classe gramatical').

E a classe gramatical depende do treinamento realizado.

```python
#Etiquetador

from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger #Etiqueta trazendo a classe gramatical mais provável, usando 1 token. Para o token que queremos etiquetar, ele vai olhar no corpus treinado e ver qual a maior probabilidade (frequencia) da classe gramatical daquele token. Há também o BigramTagger e NgramTagger também.

#tokenizar o texto
tokens = nltk.word_tokenize(corpus)

#pegar as sentenças do mac_morpho e treinar o etiquetador com elas
sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino) #recebe as sentenças de treino para treinar

etiquetado = etiquetador.tag(tokens) #o método tag etiqueta. Então passo o que quero etiquetar.

print(etiquetado)
```

Algumas etiquetas vieram como "None", pois só será possível etiquetar se a palavra estiver presente no corpus mac_morpho. 

Ex. de etiqueta 'None': Patriots, Azarões, etc.

O etiquetador não consegue inferir a classe gramatical de palavras não vistas no treinamento.

E como resolver?

Acrescentar um etiquetador padrão que dará uma classe caso a palavra não exista no texto de referência. Normalmente funciona bem.

A solução é pré-classificar todas as palavras do texto como substantivos(N) e depois treinar o etiquetador normalmente. Usa-se o pacote **DefaultTagger**.

```python
#Etiquetador com padrão DefaultTagger

from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger #Etiqueta trazendo a classe gramatical mais provável, usando 1 token. Para o token que queremos etiquetar, ele vai olhar no corpus treinado e ver qual a maior probabilidade (frequencia) da classe gramatical daquele token. Há também o BigramTagger e NgramTagger também.
from nltk.tag import DefaultTagger

#tokenizar o texto
tokens = nltk.word_tokenize(corpus)

#pegar as sentenças do mac_morpho e treinar o etiquetador com elas
etiquetador_padrao = DefaultTagger('N') #Usaremos como padrão que as palavras são substantivos
sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino, backoff=etiquetador_padrao) #backoff irá considerar como substantivo ('N'), caso não encontre a palavra no corpus de origem. 

etiquetado = etiquetador.tag(tokens) #o método tag etiqueta. Então passo o que quero etiquetar.

print(etiquetado)
```

Note que agora os tokens 'Patriots', 'Azarões' foram etiquetados como substantivos ('N').

Com as tuplas (palavras + etiquetadas), podemos fazer várias coisas:

- E várias outras.

```python
from nltk.chunk import RegexpParser

#precisamos montar um padrão para a busca, no caso, de entidades nomeadas
#NP = Noun Phrase, ou sintagma nominal: é um par de nomes próprios OU um par de susbstantivos.
pattern = 'NP: {<NPROP><NPROP> | <N><N>}' #par de nomes próprios (NPROP) ou par de substantivos (N)

analise_gramatical = RegexpParser(pattern) 

#gera uma pequena árvore (pedaços ou 'chunks'). E passamos os tokens etiquetados
arvore = analise_gramatical.parse(etiquetado)
print(arvore)
```
